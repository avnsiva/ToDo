```python
banknifty = pd.read_csv("C:\Users\sivar\Dropbox\Python Notebooks\OptionsData\\July_Expiries_withRange.csv", date_parser=True)
banknifty['Date'] = pd.to_datetime(banknifty['Date'])
banknifty['Expiry'] = pd.to_datetime(banknifty['Expiry'])
expnifty = banknifty.drop_duplicates(['Expiry'], keep='last')
expiries = expnifty['Expiry'].tolist()
DF5 = banknifty
CADF = expiries
PEDF = expiries
```

```python
for expiry in expiries:
    print expiry
    duration = expiry - datetime.timedelta(days=5)
    ExpiryDF = banknifty[banknifty['Expiry'] == expiry]
    DF5 = ExpiryDF[ExpiryDF['Date'] > duration]
    #DF5['PositionUp'] = int(math.ceil(ExpiryDF[ExpiryDF['Date'] == ExpiryDF['Date'].min()].drop_duplicates(['Underlying'], keep='last').UpperRange/100)*100)
    #DF5['PositionDown'] = int(round(ExpiryDF[ExpiryDF['Date'] == ExpiryDF['Date'].min()].drop_duplicates(['Underlying'], keep='last').LowerRange, -2))
    DF5['PositionUp'] = int(math.ceil(DF5[DF5['Date'] == DF5['Date'].min()].drop_duplicates(['Underlying'], keep='last').UpperRange/100)*100)
    DF5['PositionDown'] = int(round(DF5[DF5['Date'] == DF5['Date'].min()].drop_duplicates(['Underlying'], keep='last').LowerRange, -2))
    strikesDF = DF5[DF5['Date'] == DF5['Date'].min()]
    strikesDF = DF5[(DF5['Strike Price'] == DF5['PositionUp']) | (DF5['Strike Price'] == DF5['PositionDown'])]
    CADF = strikesDF[(strikesDF['Option Type'] == 'CE') & (strikesDF['Strike Price'] == strikesDF['PositionUp'])]
    PEDF = strikesDF[(strikesDF['Option Type'] == 'PE') & (strikesDF['Strike Price'] == strikesDF['PositionDown'])]
    print expiry.strftime('%m/%d/%Y')
    createCharts(ExpiryDF,DF5,expiry.strftime('%m/%d/%Y'), CADF,PEDF )
    time.sleep(10)
```

```python
def createCharts(ExpiryDF,DF5,fName, CA, PE):
    fname = 'C:\Users\sivar\Dropbox\Python Notebooks\OptionsData\\'+fName
    print CA['Close_x']
    #trace1 = go.Scatter(x=CA['Date'], y=CA['AnnualVolatility365'],name='Volatility')
    trace2 = go.Scatter(x=ExpiryDF['Date'], y=ExpiryDF['AnnualVolatility365'],name='Volatility')
    trace3 = go.Scatter(x=DF5['Date'], y=DF5['High_y'],name='High')
    trace4 = go.Scatter(x=DF5['Date'], y=DF5['Low_y'],name='Low')
    trace5 = go.Scatter(x=DF5['Date'], y=DF5['Close_y'],name='Close')
    trace6 = go.Scatter(x=CA['Date'], y=CA['Close_x'],name='CA CLose')
    trace7 = go.Scatter(x=PE['Date'], y=PE['Close_x'],name='PE Close')
    trace8 = go.Candlestick(x=DF5['Date'], open=DF5['Open_y'],high=DF5['High_y']
                            ,low=DF5['Low_y'],close=DF5['Close_y'], name='Candlestick')
    trace9 = go.Scatter(x=DF5['Date'], y=DF5['PositionUp'],name='PositionUp')
    trace10 = go.Scatter(x=DF5['Date'], y=DF5['PositionDown'],name='PositionDown')
    trace11 = go.Scatter(x=DF5['Date'], y=DF5['UpperRange'],name='UpperRange')
    trace12 = go.Scatter(x=DF5['Date'], y=DF5['LowerRange'],name='LowerRange')
    # data = [trace2]
    fig = tools.make_subplots(rows=3, cols=3,)
    
    #fig.append_trace(trace1, 1, 1)
    fig.append_trace(trace2, 2, 1)
    fig.append_trace(trace3, 3, 1)
    fig.append_trace(trace4, 3, 1)
    fig.append_trace(trace5, 3, 1)
    fig.append_trace(trace6, 3, 2)
    fig.append_trace(trace7, 1, 3)
    fig.append_trace(trace8, 2, 3)
    fig.append_trace(trace9, 3, 1)
    fig.append_trace(trace10, 3, 1)
    fig.append_trace(trace11, 3, 1)
    fig.append_trace(trace12, 3, 1)
    fig['layout'].update(title=fName)
    plot_url=offline.plot(fig, filename='123')
```
